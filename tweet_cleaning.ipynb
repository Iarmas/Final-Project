```python
import json
import pandas as pd
```

```python
f1 = "metrocdmx_1331_30m.json"
f2 = "metrocdmx_1527_60m_71119.json"
f3 = "metrocdmx_1605_120m.json"
f4 = "metrocdmx_1716_120m.json"
f5 = "metrocdmx_1827_60m.json"
f6 = "metrocdmx_1930_60m.json"
f7 = "metrocdmx_2100_60m.json"
f8 = "metrocdmx_841_80m.json"

file_arr = [f1, f2, f3, f4, f5, f6, f7, f8]
```

```python
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
nltk.download('perluniprops')
nltk.download('nonbreaking_prefixes')

from nltk.tokenize.toktok import ToktokTokenizer

toktok = ToktokTokenizer()
```

```python
estaciones = "Lineas_Estaciones.json"

with open(estaciones, 'r') as cat:
    cat_dict = json.loads(cat.read())

cat_keys = cat_dict.keys()
cat_dict
```

```python
tweet_dict = {
            "date" : [],
            "hour" : [],
            "tweet" : [],
            "linea" : [],
            "estacion" : []
        }

for f in file_arr:

    with open(f, 'r') as json_file:
        for line in json_file:
            cont = 0
            tweet = json.loads(line)
    
            full_date = tweet['created_at']
            full_date_split = full_date.split()
            
            hour = full_date_split[3]
            date = full_date_split[2] + "/" + full_date_split[1] + "/" + full_date_split[-1]

            if "retweeted_status" in tweet.keys():
                if "extended_tweet" in tweet['retweeted_status'].keys():
                    text = tweet['retweeted_status']['extended_tweet']['full_text']
                else:
                    text = tweet['retweeted_status']['text']
    
            elif "extended_tweet" in tweet.keys():
                text = tweet['extended_tweet']['full_text']
            else:
                text = tweet['text']
                
            text_split = toktok.tokenize(text) # text.split()
            print(text_split)

            est = ""
            linea = ""
            for c_n in cat_dict.keys():
                for word in text_split:
                    word = word.replace("#", "")
                    if word.lower() in cat_dict[c_n]['Estaciones']:
                        est += word + " "
                    if word.lower() in cat_dict[c_n]['Nombre']:
                        linea += word + " "
                if linea == "" and est != "":
                    linea += cat_dict[c_n]['Nombre'][1] + " "

            tweet_dict['date'].append(date)
            tweet_dict['hour'].append(hour)
            tweet_dict['tweet'].append(text)
            tweet_dict['linea'].append(linea)
            tweet_dict['estacion'].append(est)
#             print(est)
#             print(linea)
```

```python
tweet_df = pd.DataFrame(tweet_dict)
tweet_df.head(100)
```

```python
tweet_df.to_csv("tweets_metrocdmx.csv")
```
