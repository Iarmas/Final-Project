```python
#Extract textfields from tweets
raw_tweets = []
for tweets in user_timeline:
    raw_tweets.append(tweets['text'])
```

```python
print(json.dumps(raw_tweets, indent=4, sort_keys=True))
```

```python
#Create a string form of our list of text
raw_string = ''.join(raw_tweets)
no_links = re.sub(r'http\S+', '', raw_string)
no_unicode = re.sub(r"\\[a-z][a-z]?[0-9]+", '', no_links)
no_special_characters = re.sub('[^A-Za-z ]+', '', no_unicode)
```

```python
raw_string
```

```python
stop_words = get_stop_words('spanish')
words = no_special_characters.split(" ")
words = [w for w in words if len(w) > 2]  # ignore a, an, be, ...
words = [w.lower() for w in words]
stop_words
```

```python
not_included=['buenas','tardes','noches','días','buena','tarde','noche','viaje','trenes','movilidadcdmx']
stop_words = stop_words + not_included
words = [w for w in words if w not in stop_words]
stop_words
```

```python
mask = np.array(Image.open('/Users/ingridarmas/Downloads/metro.png'))
```

```python
wc = WordCloud(background_color="white", max_words=2000, mask=mask)
clean_string = ','.join(words)
wc.generate(clean_string)
```

```python
f = plt.figure(figsize=(50,50))
f.add_subplot(1,2, 1)
plt.imshow(mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.title('Original Stencil', size=40)
plt.axis("off")
f.add_subplot(1,2, 2)
plt.imshow(wc, interpolation='bilinear')
plt.title('Twitter Generated Cloud', size=40)
plt.axis("off")
plt.show()
```
